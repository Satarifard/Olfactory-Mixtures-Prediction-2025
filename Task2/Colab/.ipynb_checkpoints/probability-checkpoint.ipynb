### Load in all the definitions.
So that we can map stimuli labels to the SMILES of the components

```python
# For some reason the SMILES have unicode symbols. Let's fix those.
def decode_unicode_escapes(smiles: str) -> str:
    return smiles.replace('\xa0', ' ')

cid_df = pd.read_csv(os.path.join(DATA_DIR,"CID.csv"))
cid_df = cid_df[cid_df['is_solvent']=="FALSE"].drop(columns="is_solvent")
cid_df["SMILES"] = cid_df["SMILES"].apply(decode_unicode_escapes)

print("NaN Molecules:", cid_df["SMILES"].isna().sum())
display(cid_df[cid_df["SMILES"].isna()])
print()

cid_to_smiles = cid_df.set_index("molecule")["SMILES"].to_dict()
len(cid_to_smiles), next(iter(cid_to_smiles.items()))
```

Sanity check we can parse everything.

```python
import rdkit

for smiles in cid_to_smiles.values():
  mol = rdkit.Chem.MolFromSmiles(smiles, sanitize=False)
  assert mol is not None, f"{smiles} is malformed"
```

Great! Continue reading in metadata.

```python
component_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Component_definition.csv"))
component_df["SMILES"] = component_df["CID"].map(cid_to_smiles)
print("NaN Components:", component_df["SMILES"].isna().sum())
display(component_df[component_df["SMILES"].isna()])
print()
component_to_smiles = component_df.set_index("id")["SMILES"].to_dict()
type(next(iter(component_to_smiles.keys()))), len(component_to_smiles), next(iter(component_to_smiles.items()))
```

```python
stimulus_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Stimulus_definition.csv"))
stimulus_df["components"] = stimulus_df["components"].apply(lambda cs: cs.split(";"))
stimulus_df["components_SMILES"] = stimulus_df["components"].apply(lambda cs: [component_to_smiles[int(c)] for c in cs])
stimuli_to_components = stimulus_df.set_index("id")["components_SMILES"].to_dict()
len(stimuli_to_components), next(iter(stimuli_to_components.items()))
```

### Not bad! As a second pass, we could just recompute only those?

```python
from tqdm.notebook import tqdm
from ogb.utils import smiles2graph
import dream.pairdata

graph_data = dict()
errored = 0

# Compute individual graphs for all unique SMILES
for smiles in tqdm(cid_to_smiles.values(), desc="Processing SMILES to graphs"):
    try:
        graph_data[smiles] = dream.pairdata.to_torch(smiles2graph(smiles))
    except AttributeError as e:
        print(f"Error processing SMILES: {smiles}. Error: {e}")
        errored += 1
    except TypeError as e:
        print(f"Error processing SMILES: {smiles}. Error: {e}")
        errored += 1

f"Errored smiles: {errored}"
```

```python
import dream.data
import time
from torch_geometric.data import Batch
from scipy.spatial import distance
import torch


def _embed(graph,normalize=False):
  with torch.no_grad():
    embedding = graph_model(graph.to(DEVICE))["embed"].cpu().squeeze()

  if normalize:
    return embedding/torch.norm(embedding,dim=-1,keepdim=True)

  return embedding

def make_embedding(blend_smiles,normalize=False):
  assert len(blend_smiles) == len(set(blend_smiles)), blend_smiles
  # For consistency, sort the smiles in the blend.
  blend_smiles = sorted(blend_smiles)
  graphs = [graph_data[smiles] for smiles in blend_smiles]
  blend_data = dream.data.combine_graphs(graphs)
  return _embed(blend_data,normalize)

ex_components = train_data_df["components_SMILES"][0]
ex_components, make_embedding(ex_components).shape
```

```python
tqdm.pandas(smoothing=0)
train_data_df["embed"] = train_data_df["components_SMILES"].progress_apply(make_embedding)
```

Let's also stack the descriptors into a single array

```python
import numpy as np
# Let's also stack y
train_data_df["descs"] = train_data_df.apply(lambda r: np.array(r[desc_cols]).astype(float), axis=1)
train_data_df["descs"][0].shape
```

Let's do a quick 3-fold validation.

```python
import sklearn
import sklearn.model_selection

import sklearn.linear_model

from scipy.spatial.distance import cosine
from scipy.stats import pearsonr
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

def get_xy(df):
    return np.stack(df["embed"].values), np.stack(df["descs"].values)

def to_binary(arr):
    return (arr > 0).astype(int)

def score(y_true, y_pred):
    cosine_dist = np.mean([cosine(u, v) for u, v in zip(y_true, y_pred)])
    pearsons = np.mean([pearsonr(u, v)[0] for u, v in zip(y_true, y_pred)])
    return [cosine_dist, pearsons]

N_SPLITS = 5
STATE = 42

sss = sklearn.model_selection.ShuffleSplit(n_splits=N_SPLITS, random_state=STATE, test_size=.25)
scores = []

for train_index, test_index in tqdm(sss.split(train_data_df), total=N_SPLITS):
    train_df, test_df = train_data_df.iloc[train_index], train_data_df.iloc[test_index]

    train_x, train_y = get_xy(train_df)
    test_x, test_y = get_xy(test_df)

    prob_model = OneVsRestClassifier(LogisticRegression(class_weight='balanced', max_iter=1000, solver='lbfgs', random_state=42)).fit(train_x, to_binary(train_y))
    score_model = LinearRegression(positive=True).fit(train_x,train_y)

    test_pred = prob_model.predict(test_x) * score_model.predict(test_x)
    
    scores.append(score(test_y, test_pred))

scores = np.stack(scores).mean(axis=0)
print(f"({scores[0]:.3f},{scores[1]:.3f})")
```
