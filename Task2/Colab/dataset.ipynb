### Load in all the definitions.
So that we can map stimuli labels to the SMILES of the components

```python
# For some reason the SMILES have unicode symbols. Let's fix those.
def decode_unicode_escapes(smiles: str) -> str:
    return smiles.replace('\xa0', ' ')

cid_df = pd.read_csv(os.path.join(DATA_DIR,"CID.csv"))
cid_df = cid_df[cid_df['is_solvent']=="FALSE"].drop(columns="is_solvent")
cid_df["SMILES"] = cid_df["SMILES"].apply(decode_unicode_escapes)

print("NaN Molecules:", cid_df["SMILES"].isna().sum())
display(cid_df[cid_df["SMILES"].isna()])
print()

cid_to_smiles = cid_df.set_index("molecule")["SMILES"].to_dict()
len(cid_to_smiles), next(iter(cid_to_smiles.items()))
```

Sanity check we can parse everything.

```python
import rdkit

for smiles in cid_to_smiles.values():
    mol = rdkit.Chem.MolFromSmiles(smiles, sanitize=False)
    assert mol is not None, f"{smiles} is malformed"
```

Great! Continue reading in metadata.

```python
component_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Components_definition.csv"))
component_df["SMILES"] = component_df["molecule"].map(cid_to_smiles)
print("NaN Components:", component_df["SMILES"].isna().sum())
display(component_df[component_df["SMILES"].isna()])
print()
component_to_smiles = component_df.set_index("id")["SMILES"].to_dict()
type(next(iter(component_to_smiles.keys()))), len(component_to_smiles), next(iter(component_to_smiles.items()))
```

```python
stimulus_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Stimulus_definition.csv"))
stimulus_df["components"] = stimulus_df["components"].apply(lambda cs: cs.split(";"))
stimulus_df["components_SMILES"] = stimulus_df["components"].apply(lambda cs: [component_to_smiles[int(c)] for c in cs])
stimuli_to_components = stimulus_df.set_index("id")["components_SMILES"].to_dict()
len(stimuli_to_components), next(iter(stimuli_to_components.items()))
```

### Load in training data.

```python
first_data_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Train_mixture_Dataset.csv"))
leaderboard_data_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Leaderboard_ActualValue.csv"))
single_data_df = pd.read_csv(os.path.join(DATA_DIR,"Task2_single_RATA.csv"))
train_data_df = pd.concat([first_data_df, leaderboard_data_df, single_data_df], axis=0).reset_index(drop=True)
```

```python
train_data_df["components_SMILES"] = train_data_df["stimulus"].map(stimuli_to_components)
train_data_df
```

```python
train_data_df = train_data_df.drop(columns=['components', 'molecule', 'dilution','NoOdor']).dropna(subset="components_SMILES")
train_data_df
```

### Temporary Fix: Drop the columns for which we don't all the smiles

```python
train_data_df["is_invalid"] = train_data_df["components_SMILES"].apply(lambda cs: any([pd.isna(s) for s in cs]))
train_data_df[train_data_df["is_invalid"]]
```

```python
train_data_df = train_data_df[~train_data_df["is_invalid"]].copy()
len(train_data_df)
```

### Let's also drop Pleasantness and Intensity

```python
train_data_df = train_data_df.drop(columns=["Intensity","Pleasantness","is_invalid"])
train_data_df
```

### Load Test Data

```python
test_form_df = pd.read_csv(os.path.join(DATA_DIR,"TASK2_Test_set_Submission_form.csv"))
raw_test_form = test_form_df.copy()

test_form_df["components_SMILES"] = test_form_df["stimulus"].map(stimuli_to_components)
test_form_df
```
